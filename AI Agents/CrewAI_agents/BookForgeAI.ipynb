### Chapter-Word Count Module

```python
## Agent 
chapter_word_agent = Agent(role="Word Count Strategist",
                          goal="Based on the Given Outline : {outline} , Synopsis : {synopsis} and Title : {title} , "
                               "Devise an optimal word count distribution across chapters, "
                               "ensuring alignment with the outline and content significance, " 
                               "while staying within the total word count limit of {total_word_count} words.",
                          backstory="You are a meticulous strategist with a deep understanding of narrative structures and pacing. " 
                                    "With an analytical mind, you assess the importance and complexity of each chapter based on the provided Outline and Synopsis. "
                                    "Your ability to estimate how much weight each chapter should carry ensures that the book's flow feels balanced, engaging, and natural. " 
                                    "You understand that climactic or pivotal chapters often need more detail, while transitional sections might require brevity. " 
                                    "Your work guarantees the book's word count is utilized efficiently, creating an engaging and satisfying reader experience.",
                          verbose=True,
                          allow_delegation=False,
                          llm=llm)

## Task
chapter_word_task = Task(description= "Analyze the provided book outline and distribute the total word count of {total_word_count} words, across chapters based on their significance and complexity."
                                     " Consider factors such as key plot points, character development, narrative pacing, and transitions. "
                                     "Allocate more words to pivotal chapters while keeping transitional chapters concise. "
                                     "Provide reasoning for each word count allocation to justify your decisions. ",
                        expected_output="A detailed list showing the word count for each chapter, along with a brief justification for the word allocation."
                                    " Each justification should reference the chapter's importance, role in the story, and complexity based on the outline. "
                                    "Ensure the total word count matches the specified limit.",
                        agent=chapter_word_agent,
                        output_pydantic=ChapterCount)
```

### Chapter components module

```python
class ChapterComponents(BaseModel):
    chapter_components : dict = Field(...,description="Dictionary of Dictionary chatper names along with their components as keys and its description as value "
                                      "ex: {'name_of_chapter':{'Role':'','Key Events':'','Tone and Mood':'','Characters and Goals/focus points':'','Transition':''},}"
                                                     "Only use chapter names as keys Don't use any other names.")
    
```

```python
chapter_components_agent = Agent(
    role="Chapter Architect",
    goal="Based on the provided Outline : {outline}, Synopsis : {synopsis}, and Title : {title} and this is the chapter list : {chapter_list}, "
    "craft the key components for each chapter, including the Role, Key Events, Tone and Mood, Characters and Goals, "
    "and ensure a clear Transition between chapters. Ensure each chapter's elements align with the overall narrative and tone of the book.",
    backstory="You are a skilled storyteller with a deep understanding of narrative design. "
    "Your role is to break down each chapter into its most essential components, ensuring that the chapter serves its unique purpose within the book. "
    "You pay special attention to the key events, tone, and mood to ensure they align with the overall narrative flow. "
    "You understand that characters’ goals and transitions between chapters are critical for maintaining a cohesive and engaging story.",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

chapter_components_task = Task(
    description="Analyze the provided book outline and determine the key components for each chapter, including:"
    " - Role: Define the chapter's purpose and contribution to the overall narrative."
    " - Key Events: Extract the critical events or moments that drive the plot forward in the chapter."
    " - Tone and Mood: Match the tone and mood of the chapter with the overarching theme and synopsis of the book."
    " - Characters and Goals (for fiction): Identify which characters are featured in the chapter and define their goals and motivations."
    " - Transition: Ensure a clear and logical transition between this chapter and the next, maintaining the book's flow.",
    expected_output="For each chapter, provide the following key components:"
    " 1. Role: A clear description of the chapter's purpose in the context of the book's plot."
    " 2. Key Events: A list of significant events or actions in the chapter that influence the story."
    " 3. Tone and Mood: A description of the emotional atmosphere and tone of the chapter."
    " 4. Characters and Goals (for fiction): A summary of the characters involved and their motivations, or focus points (for non-fiction)."
    " 5. Transition: A brief explanation of how the chapter transitions into the next chapter, ensuring smooth narrative progression.",
    agent=chapter_components_agent,
    output_pydantic=ChapterComponents
)
```

```python
crew_n = Crew(agents=[chapter_components_agent],
           tasks=[chapter_components_task],
           verbose=True)

inputs = {'title':title_spanish,
         'synopsis':synopsis_spanish,
         'outline':outline_spanish,
         'chapter_list':chapter_list}
```
